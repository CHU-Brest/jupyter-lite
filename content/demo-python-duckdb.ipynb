```python
import duckdb
import pandas as pd

print("duckdb", duckdb.__version__)
print("pandas", pd.__version__)
```

```python
# Un DataFrame pandas
df = pd.DataFrame(
    {
        "ville": ["Brest", "Brest", "Rennes", "Rennes"],
        "annee": [2023, 2024, 2023, 2024],
        "valeur": [10, 15, 7, 21],
    }
)
df
```

```python
# Interroger directement le DataFrame pandas avec DuckDB (SQL)
duckdb.sql(
    """
    SELECT ville, SUM(valeur) AS total
    FROM df
    GROUP BY ville
    ORDER BY total DESC
    """
).df()
```
