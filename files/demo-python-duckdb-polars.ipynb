```python
import duckdb
import polars as pl

print("duckdb", duckdb.__version__)
print("polars", pl.__version__)
```

```python
# Un DataFrame Polars
df = pl.DataFrame(
    {
        "ville": ["Brest", "Brest", "Rennes", "Rennes"],
        "annee": [2023, 2024, 2023, 2024],
        "valeur": [10, 15, 7, 21],
    }
)
df
```

```python
# Interroger directement le DataFrame Polars avec DuckDB (SQL)
duckdb.sql(
    """
    SELECT ville, SUM(valeur) AS total
    FROM df
    GROUP BY ville
    ORDER BY total DESC
    """
).pl()
```
